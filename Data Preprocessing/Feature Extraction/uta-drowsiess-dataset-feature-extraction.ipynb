```python
def removealpha(string, max_char,):
    string = string[-6:]
    temp = string[0:max_char+1]
    result = []
    for c in temp:
        if c.isnumeric():
            result.append(c)
        
    return int(''.join(result))
```

```python
def eye_aspect_ratio(eye):
    A = distance.euclidean(eye[1], eye[5])
    B = distance.euclidean(eye[2], eye[4])
    C = distance.euclidean(eye[0], eye[3]) 
    ear = (A + B) / (2.0 * C)
    return ear
def mouth_aspect_ratio(mouth):
    A = distance.euclidean(mouth[14], mouth[18])
    C = distance.euclidean(mouth[12], mouth[16])
    mar = (A ) / (C)
    return mar

def circularity(eye):
    radius = distance.euclidean(eye[1], eye[4]) / 2
    area = math.pi * (radius ** 2)
    perimeter = sum(distance.euclidean(eye[i], eye[i+1]) for i in range(5))
    return 4 * math.pi * area / (perimeter ** 2)

def mouth_over_eye(eye):
    ear = eye_aspect_ratio(eye)
    mar = mouth_aspect_ratio(eye)
    mouth_eye = mar/ear
    return mouth_eye

def getFrame(sec,vidcap):
    vidcap.set(cv2.CAP_PROP_POS_MSEC, 1800 + sec*1000)
    return vidcap.read()
```

```python
def feature_extraction(path, state, ID, N = 300, frame_rate = 1):
    #path is string to video
    #state : 0 , 5 , 10
    data = []
    features = []
    vidcap = cv2.VideoCapture (path)
    sec = 0
    success, image  = getFrame(sec,vidcap)
    count = 0
    #print( success)
    #print(image)
    while success and count < N:
        landmarks = extract_face_landmarks(image)
        if landmarks is not None:
            
            count += 1
            data.append(landmarks)
            sec = sec + frame_rate
            sec = round(sec, 2)
            success, image = getFrame(sec,vidcap)
            print(count)
            
            
        else:
            sec = sec + frame_rate
            sec = round(sec, 2)
            success, image = getFrame(sec,vidcap)
            print("not detected")
    
    # Calculate features from landmarks
    
    for x in data:
        shape = x[36:68]
        ear = eye_aspect_ratio(shape)
        mar = mouth_aspect_ratio(shape)
        cir = circularity(shape)
        moe = mouth_over_eye(shape)
        features.append({'ID': ID,'EAR': ear,'MAR': mar,'CIR':cir,'MOE': moe,'State':state})
        
    return features
```
