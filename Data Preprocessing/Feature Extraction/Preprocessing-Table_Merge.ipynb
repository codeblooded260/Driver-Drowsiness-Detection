```python
import pandas as pd
import os
import glob
```

```python
tables = glob.glob("..\\Data\\*.csv")
```

```python
tables
```

```python
pid = 0
cols_name = ['ID','EAR','MAR','CIR','MOE','State']
for i in range(0,len(tables)-1,3):
    print(i)
    table_1 = f"..\\Data\\Table_{i}.csv"
    table_2 = f"..\\Data\\Table_{i+1}.csv"
    table_3 = f"..\\Data\\Table_{i+2}.csv"
    df_1 = pd.read_csv(table_1,header = None, names=cols_name)
    df_2 = pd.read_csv(table_2,header = None, names=cols_name)
    df_3 = pd.read_csv(table_3,header = None, names=cols_name)
    
    df_merged = pd.concat([df_1,df_2,df_3], axis=0)
    df_merged.to_csv(f"..\\Merged\\Table_{pid}.csv")
    pid += 1
```
